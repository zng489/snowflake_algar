```
WITH


FAT_CNPJ AS (
    SELECT 
        distinct
        VPBC.RAIZ_CNPJ,
        VPBC.CNPJ_CPF,
        SUM(VPBC.FATURAMENTO) AS FATURAMENTO_GROUP_BY
    FROM DWDEV.DWADM.V_PBI_BASE_CORPORATIVO_CONTRATO AS VPBC
    WHERE VPBC.MES_ANO = '202601'
      AND VPBC.INDICADOR = 'Base'
      AND VPBC.STATUS = 'ATIVO'
      AND VPBC.SEGMENTO = 'CORPORATIVO'
    GROUP BY VPBC.RAIZ_CNPJ,CNPJ_CPF
),


CNPJ_PRINCIPAL AS (
    SELECT distinct
        RAIZ_CNPJ,
        CNPJ_CPF,
        FATURAMENTO_GROUP_BY,
        FIRST_VALUE(CNPJ_CPF) OVER (
    PARTITION BY RAIZ_CNPJ
        ORDER BY FATURAMENTO_GROUP_BY DESC, CNPJ_CPF
    ) AS CNPJ_PRINCIPAL
FROM FAT_CNPJ),

FAT AS (
    SELECT 
        distinct
        VPBC.RAIZ_CNPJ,
        --VPBC.CNPJ_CPF,
        SUM(VPBC.FATURAMENTO) AS FATURAMENTO_GROUP_BY
    FROM DWDEV.DWADM.V_PBI_BASE_CORPORATIVO_CONTRATO AS VPBC
    WHERE VPBC.MES_ANO = '202601'
      AND VPBC.INDICADOR = 'Base'
      AND VPBC.STATUS = 'ATIVO'
      AND VPBC.SEGMENTO = 'CORPORATIVO'
    GROUP BY VPBC.RAIZ_CNPJ
),

FAIXA AS (
    SELECT DISTINCT
        RAIZ_CNPJ,
        FATURAMENTO_GROUP_BY,
        CASE
            WHEN FATURAMENTO_GROUP_BY <= 2499.99 THEN 'Até R$2.499,99'
            WHEN FATURAMENTO_GROUP_BY BETWEEN 2500 AND 4999.99 THEN 'R$2.500,00 a R$4.999,99'
            WHEN FATURAMENTO_GROUP_BY BETWEEN 5000 AND 9999.99 THEN 'R$5.000,00 a R$9.999,99'
            WHEN FATURAMENTO_GROUP_BY BETWEEN 10000 AND 24999.99 THEN 'R$10.000,00 a R$24.999,99'
            WHEN FATURAMENTO_GROUP_BY BETWEEN 25000 AND 49999.99 THEN 'R$25.000,00 a R$49.999,99'
            WHEN FATURAMENTO_GROUP_BY BETWEEN 50000 AND 79999.99 THEN 'R$50.000,00 a R$79.999,99'
            WHEN FATURAMENTO_GROUP_BY BETWEEN 80000 AND 99999.99 THEN 'R$80.000,00 a R$99.999,99'
            ELSE 'Acima de R$100.000,00'
        END AS FAIXA_FATURAMENTO,

    FROM FAT
),

-- TEM QUE VIR ANTES DO FINAL
CONTRACT_KEY_SUB_PILARES_LISTA AS (
    SELECT DISTINCT
        RAIZ_CNPJ,
        COUNT(DISTINCT CONTRACT_KEY) AS VOL_CONTRACTS,
        COUNT(DISTINCT SUB_PILARES) AS CESTA,
        LISTAGG(
            DISTINCT UPPER(REPLACE(SUB_PILARES,' ','_')),
            ', '
        ) AS LISTA_SUBPILARES
    FROM DWDEV.DWADM.V_PBI_BASE_CORPORATIVO_CONTRATO
    GROUP BY RAIZ_CNPJ
),

FINAL AS (
    SELECT DISTINCT
        F.RAIZ_CNPJ,
        F.FATURAMENTO_GROUP_BY as FATURAMENTO_RAIZ,
        F.FAIXA_FATURAMENTO,
        S.VOL_CONTRACTS,
        S.CESTA,
        S.LISTA_SUBPILARES
    FROM FAIXA F
    LEFT JOIN CONTRACT_KEY_SUB_PILARES_LISTA S
        ON F.RAIZ_CNPJ = S.RAIZ_CNPJ
),


-- DWADM.V_PBI_BASE_CORPORATIVO_CONTRATO
V_PBI_BASE_CORPORATIVO_CONTRATO_1 AS (
    SELECT distinct
        VPBC.MES_ANO,
        VPBC.INDICADOR,
        VPBC.STATUS,
        VPBC.SEGMENTO,
        VPBC.CONTRACT_KEY,
        VPBC.CUSTOMER_KEY,
        VPBC.PRODUCT_KEY,
        VPBC.CLIENTE,
        VPBC.RAIZ_CNPJ,
        LPAD(VPBC.CNPJ_CPF, 14, '0') AS CNPJ_CPF,
        VPBC.TPO_END,
        VPBC.TPO_RUA,
        VPBC.RUA,
        VPBC.NRO,
        VPBC.COMPLEMENT,
        VPBC.BAIRRO,
        VPBC.CIDADE,
        VPBC.ESTADO,
        VPBC.LOCALIDADE_UF,
        VPBC.CEP,
        VPBC.FLG_TIP_CLI_CNPJ_CPE,
        VPBC.FLG_TIP_CLI_CNPJ_RAIZ,
        VPBC.FLG_CLI_PRM_GTD,
        VPBC.CONSULTOR,
        VPBC.TIPO_CONSULTOR,
        VPBC.DIRETORIA,
        VPBC.DESC_REG_CONS,
        VPBC.REGIONAL_CONSULTOR,
        VPBC.REGIONAL_CONTRATO,
        VPBC.DESC_REGIONAL_CONTRATO,
        VPBC.REGIONAL_GESTAO_EBIT,
        VPBC.DESC_REGIONAL_GESTAO_EBIT,
        VPBC.REGIONAL_LOCALIZACAO,
        VPBC.DESC_REGIONAL_LOCALIZACAO,
        VPBC.LOCALIDADE_UF_NEOWAY,
        VPBC.AREA_ATUACAO,
        VPBC.SEGMENTO_MERCADO,
        VPBC.SUBSEGMENTO_MERCADO,
        VPBC.TIER,
        VPBC.TIPO_REDE,
        VPBC.FLAG_MATRIZ,
        VPBC.SUBSCRICAO,
        VPBC.ORIGEM,
        VPBC.PRODUTO,
        VPBC.CIRCUITO,
        VPBC.PILARES,
        VPBC.SUB_PILARES,
        VPBC.VLC_RAMAIS,
        VPBC.PROD_DEMANDA,
        VPBC.LINHA_DE_NEGOCIO,
        VPBC.QTD_TNC_SMO,
        VPBC.DSC_TCN,
        VPBC.FLG_PDT_TIC,
        VPBC.RELEVANCIA_ESTRATEGICA,
        VPBC.RISCO,
        -- Uma parte
        VPBC.FATURAMENTO,
        VPBC.DATA_ATV_CONTRATO_ATV_ANTIGO,
        YEAR(VPBC.DATA_ATV_CONTRATO_ATV_ANTIGO) AS ANO,
        -- 1 anos desde ativação
        DATEDIFF(YEAR, DATA_ATV_CONTRATO_ATV_ANTIGO, CURRENT_DATE) 
            AS ANOS_DESDE_ATIVACAO,
        -- 2 dia/mês
        TO_CHAR(DATA_ATV_CONTRATO_ATV_ANTIGO, 'DD/MM') 
            AS DIA_MES_ANIVERSARIO,
        -- 3 próxima data
        CASE 
            WHEN DATE_FROM_PARTS(
                    YEAR(CURRENT_DATE),
                    MONTH(DATA_ATV_CONTRATO_ATV_ANTIGO),
                    DAY(DATA_ATV_CONTRATO_ATV_ANTIGO)
                 ) >= CURRENT_DATE
            THEN DATE_FROM_PARTS(
                    YEAR(CURRENT_DATE),
                    MONTH(DATA_ATV_CONTRATO_ATV_ANTIGO),
                    DAY(DATA_ATV_CONTRATO_ATV_ANTIGO)
                 )
            ELSE DATE_FROM_PARTS(
                    YEAR(CURRENT_DATE) + 1,
                    MONTH(DATA_ATV_CONTRATO_ATV_ANTIGO),
                    DAY(DATA_ATV_CONTRATO_ATV_ANTIGO)
                 )
        END AS PROXIMO_ANIVERSARIO,    
        CASE
            WHEN UPPER(SUB_PILARES) IN (
                'CSP 12',
                'LP DADOS',
                'LP VOZ',
                'PABX VIRTUAL',
                'PASOO',
                'PLANOS FIXA',
                'PLANOS MOVEL',
                'TRIDIGITO',
                'VIDEOCONFERENCIA'
            ) THEN 'SIM'
            WHEN UPPER(SUB_PILARES) IN (
                'NAO CLASSIFICADO',
                'OUTROS',
                'TV',
                'AZURE',
                'ENERGIA',
                'IOT - CONECTIVIDADE',
                'PRESENCA DIGITAL',
                'SAUDE VOUCHER',
                '500'
            ) THEN 'NAO'
            ELSE 'NAO'  -- default: não elegível
        END AS FLG_ELEGIVEL_FIDELIDADE
    FROM DWDEV.DWADM.V_PBI_BASE_CORPORATIVO_CONTRATO AS VPBC
    WHERE VPBC.MES_ANO = '202601'
      AND VPBC.INDICADOR = 'Base'
      AND VPBC.STATUS = 'ATIVO'
      AND VPBC.SEGMENTO = 'CORPORATIVO'
),


-- DWDEV.DBMADM.DBM_CARTEIRA_CLIENTE
DBM_CARTEIRA_CLIENTE AS (
    SELECT LPAD(CPF_CNPJ, 14, '0') AS CPF_CNPJ, GESTOR_CANAIS, CONGLOMERADO_RAIZ AS CONGLOMERADO, CUSTOMER_SUCESS
    FROM DWDEV.DBMADM.DBM_CARTEIRA_CLIENTE
    where MES_REFERENCIA >= DATEADD(month, -1, CURRENT_TIMESTAMP)),


-- DWDEV.DWADM.V_PBI_SEMAFORO_CORPORATIVO
V_PBI_SEMAFORO_CORPORATIVO AS (
    SELECT distinct
        CONTRACT_KEY_DW,
        VISITAS_3M AS QTD_VISITA_3M,
        VAR5_VISITA_3_M AS VISITA_3M,
        STATUS_FIDELIZACAO AS FIDELIZACAO_CONTRATO,
        YEAR(FIM_FIDELIZACAO) AS ANO_VENCIMENTO,

        CASE 
            WHEN FLAG_ACIMA_MERCADO = 'ABAIXO' THEN 0
            WHEN FLAG_ACIMA_MERCADO = 'IGUAL' THEN 1
            WHEN FLAG_ACIMA_MERCADO = 'ACIMA' THEN 2
        END AS POSICIONAMENTO,

        CASE
            WHEN VAR12_COMPORTAMENTO_DE_BASE = 4 THEN 'NOVO'
            WHEN VAR12_COMPORTAMENTO_DE_BASE < 1 THEN 'CANCELANDO'
            WHEN VAR12_COMPORTAMENTO_DE_BASE = 1 THEN 'MANTENDO'
            WHEN VAR12_COMPORTAMENTO_DE_BASE > 1 THEN 'CRESCENDO'
        END AS COMPORTAMENTO_BASE,

        QTD_ORD_SOM_CIR_6MESES AS INCIDENTES,
        TEC_IRR_30DIAS_ALGAR AS IRR,
        FLG_TROCA_CARTEIRA_6M AS TROCA_DE_CARTEIRA,
        QTD_CONTESTACOES_6M AS CONTESTACAO,
        QTD_ORD_SOM_MASSIVA_6MESES AS MASSIVA,

        CASE
            WHEN VAR10_INADIMPLENCIA > 0 THEN 'SIM'
            WHEN VAR10_INADIMPLENCIA = 0 THEN 'NÃO'
        END AS INADIMPLENCIA

    FROM DWDEV.DWADM.V_PBI_SEMAFORO_CORPORATIVO
),


-- DWDEV.DWADM.V_PBI_GEOMARKETING
V_PBI_GEOMARKETING AS (
    SELECT DISTINCT CNPJ, VERTICAL_NEGOCIO AS VERTICAL
    FROM DWDEV.DWADM.V_PBI_GEOMARKETING),



-- DWDEV.DWADM.IND_SAFRA
IND_SAFRA AS (

SELECT
    DAT_REF AS DATA_ULTIMA_CAMPANHA,
    DSC_CAM AS NOME_CAMPANHA,
    SUBSTR(NUM_CPF_CNPJ, 1, 8) AS RAIZ_CNPJ,
    CASE 
        WHEN SUBSTR(NUM_CPF_CNPJ, 1, 8) IS NOT NULL 
             AND TRIM(SUBSTR(NUM_CPF_CNPJ, 1, 8)) <> '' 
        THEN 1 
        ELSE 0 
    END AS LEAD,
    CASE 
        WHEN SUBSTR(NUM_CPF_CNPJ, 1, 8) IS NOT NULL 
             AND TRIM(SUBSTR(NUM_CPF_CNPJ, 1, 8)) <> '' 
        THEN 'SIM' 
        ELSE 'NAO' 
    END AS SIM_OU_NAO_CPF_CNPJ
FROM DWDEV.DWADM.IND_SAFRA
WHERE DSC_IND = 'AGENDAMENTO'
  AND DAT_REF >= DATEADD(day, -90, CURRENT_TIMESTAMP())
QUALIFY ROW_NUMBER() OVER (
    PARTITION BY SUBSTR(NUM_CPF_CNPJ, 1, 8)
    ORDER BY DAT_REF DESC, DSC_CAM DESC
) = 1
),


-- DWDEV.DWADM.D_OPPORTUNITY
D_OPPORTUNITY AS (
SELECT
    o.CUSTOMER_KEY,
    --o.GROUP_SEGMENT,
    o.CREATION_DATE,
    CASE 
        WHEN v.CUSTOMER_KEY IS NOT NULL THEN 1
        ELSE 0
    END AS FLG_D_OPPORTUNITY,
    v.CNPJ_CPF
FROM DWDEV.DWADM.D_OPPORTUNITY o
LEFT JOIN DWDEV.DWADM.V_PBI_BASE_CORPORATIVO_CONTRATO v
    ON o.CUSTOMER_KEY = v.CUSTOMER_KEY
WHERE o.GROUP_SEGMENT = 'CORPORATIVO'
  AND o.CREATION_DATE >= CURRENT_DATE - 90
  AND v.CUSTOMER_KEY IS NULL),

  
-- DWDEV.DWADM.D_COTACAO
COTACAO AS (
SELECT
   cc.CUSTOMER_KEY,
   cc.CNPJ_CPF,
   dc.SRK_CLI AS COTACAO,
   CASE 
       WHEN dc.SRK_CLI IS NOT NULL THEN 1
       ELSE 0
   END AS FLG_COTACAO
FROM DWDEV.DWADM.V_PBI_BASE_CORPORATIVO_CONTRATO AS cc
LEFT JOIN DWDEV.DWADM.D_COTACAO AS dc
   ON cc.CUSTOMER_KEY = dc.SRK_CLI
  AND dc.DAT_CRI_RGT >= CURRENT_DATE - 90
WHERE dc.SRK_CLI IS NULL
)


SELECT distinct
    V_PBI_BASE_CORPORATIVO_CONTRATO_1.*,
    DBM_CARTEIRA_CLIENTE.GESTOR_CANAIS,
   DBM_CARTEIRA_CLIENTE.CONGLOMERADO,
   DBM_CARTEIRA_CLIENTE.CUSTOMER_SUCESS,
   V_PBI_GEOMARKETING.VERTICAL,
   IND_SAFRA.NOME_CAMPANHA,
   IND_SAFRA.LEAD,
   FINAL.*,
   CNPJ_PRINCIPAL.FATURAMENTO_GROUP_BY AS FATURAMENTO_CNPJ,
CNPJ_PRINCIPAL.CNPJ_PRINCIPAL
FROM V_PBI_BASE_CORPORATIVO_CONTRATO_1

LEFT JOIN FAIXA 
  ON V_PBI_BASE_CORPORATIVO_CONTRATO_1.RAIZ_CNPJ = FAIXA.RAIZ_CNPJ

LEFT JOIN CNPJ_PRINCIPAL 
  ON V_PBI_BASE_CORPORATIVO_CONTRATO_1.CNPJ_CPF = CNPJ_PRINCIPAL.CNPJ_CPF

LEFT JOIN DBM_CARTEIRA_CLIENTE
   ON V_PBI_BASE_CORPORATIVO_CONTRATO_1.CNPJ_CPF = DBM_CARTEIRA_CLIENTE.CPF_CNPJ

LEFT JOIN V_PBI_SEMAFORO_CORPORATIVO
    ON V_PBI_BASE_CORPORATIVO_CONTRATO_1.CONTRACT_KEY = V_PBI_SEMAFORO_CORPORATIVO.CONTRACT_KEY_DW --AND V_PBI_BASE_CORPORATIVO_CONTRATO_1.RAIZ_CNPJ= V_PBI_SEMAFORO_CORPORATIVO.RAIZ_CNPJ

-- join geomarketing (por RAIZ)
LEFT JOIN V_PBI_GEOMARKETING
   ON V_PBI_BASE_CORPORATIVO_CONTRATO_1.CNPJ_CPF = V_PBI_GEOMARKETING.CNPJ 

-- join safra (por RAIZ)
LEFT JOIN IND_SAFRA
   ON V_PBI_BASE_CORPORATIVO_CONTRATO_1.RAIZ_CNPJ = IND_SAFRA.RAIZ_CNPJ

LEFT JOIN FINAL
    ON FINAL.RAIZ_CNPJ = V_PBI_BASE_CORPORATIVO_CONTRATO_1.RAIZ_CNPJ

LEFT JOIN D_OPPORTUNITY
    ON D_OPPORTUNITY.CNPJ_CPF= V_PBI_BASE_CORPORATIVO_CONTRATO_1.CNPJ_CPF

LEFT JOIN COTACAO
    ON COTACAO.CNPJ_CPF= V_PBI_BASE_CORPORATIVO_CONTRATO_1.CNPJ_CPF
    
WHERE IND_SAFRA.NOME_CAMPANHA LIKE '%DRE%' OR IND_SAFRA.NOME_CAMPANHA IS NULL

LIMIT 50;
```

```
WITH IND_SAFRA_BASE AS (

    SELECT
        DAT_REF,
        DSC_CAM,
        SUBSTR(NUM_CPF_CNPJ, 1, 8) AS RAIZ_CNPJ
    FROM DWDEV.DWADM.IND_SAFRA
    WHERE DSC_IND = 'AGENDAMENTO'
      AND DAT_REF >= DATEADD(DAY, -90, CURRENT_TIMESTAMP())

),

ULTIMA_DATA AS (

    SELECT
        RAIZ_CNPJ,
        MAX(DAT_REF) AS MAX_DAT_REF
    FROM IND_SAFRA_BASE
    GROUP BY RAIZ_CNPJ

),

ULTIMA_CAMPANHA AS (

    SELECT
        B.RAIZ_CNPJ,
        B.DAT_REF,
        MAX(B.DSC_CAM) AS MAX_DSC_CAM
    FROM IND_SAFRA_BASE B
    INNER JOIN ULTIMA_DATA U
        ON B.RAIZ_CNPJ = U.RAIZ_CNPJ
       AND B.DAT_REF = U.MAX_DAT_REF
    GROUP BY B.RAIZ_CNPJ, B.DAT_REF

),

IND_SAFRA AS (

    SELECT
        B.DAT_REF AS DATA_ULTIMA_CAMPANHA,
        B.DSC_CAM AS NOME_CAMPANHA,
        B.RAIZ_CNPJ,
        IFF(TRIM(B.RAIZ_CNPJ) <> '', 1, 0) AS LEAD,
        IFF(TRIM(B.RAIZ_CNPJ) <> '', 'SIM', 'NAO') AS SIM_OU_NAO_CPF_CNPJ

    FROM IND_SAFRA_BASE B
    INNER JOIN ULTIMA_CAMPANHA U
        ON B.RAIZ_CNPJ = U.RAIZ_CNPJ
       AND B.DAT_REF   = U.DAT_REF
       AND B.DSC_CAM   = U.MAX_DSC_CAM

)

SELECT * FROM IND_SAFRA
LIMIT 50;
```
